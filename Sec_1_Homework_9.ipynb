# 3.) Check a range of k-clusters and visualize to find the elbow. Test 30 different random starting places for the centroid means


```python
WCSSs = []
Ks = range(1,15)
for k in Ks:
  Kmeans = KMeans(n_clusters = k, n_init=30).fit(X_scaled)
  WCSSs.append(Kmeans.inertia_)
```

# 4.) Use the above work and economic critical thinking to choose a number of clusters. Explain why you chose the number of clusters and fit a model accordingly.

```python
plt.plot(WCSSs)
plt.show()
```

Based on the analysis of the preceding graph and the plot displayed above, it appears that selecting three clusters yields the optimal outcome. A principal justification for this conclusion lies in the observation that the Within-Cluster Sum of Squares (WCSS) enters a phase of relative stability and consistency after the establishment of three clusters. This period of reduced fluctuation indicates that the marginal advantage of incorporating additional clusters beyond three begins to decrease, thus signifying a point of diminishing returns. Therefore, a configuration of three clusters strikes an effective balance between achieving a compact clustering solution and maintaining the model's complexity at a manageable level.

# 6.) Do the same for a silhoutte plot

```python
from sklearn.metrics import silhouette_score
```

```python
SSs = []
Ks = range(2,15)
for k in Ks:
  Kmeans = KMeans(n_clusters = k, n_init=30).fit(X_scaled)
  sil = silhouette_score(X_scaled, Kmeans.labels_)
  SSs.append(sil)
```

```python
plt.plot(SSs)
plt.xlabel("# of clusters")
plt.ylabel('SSs')
plt.show()
```
